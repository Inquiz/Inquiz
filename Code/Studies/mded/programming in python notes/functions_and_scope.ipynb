# Python has different types of Scope

    Local
    Enclosing
    Global
    Built-in



The purpose of scope is to protect the variable


```python
#Built-in Scope (def)
def outer():
    #Enclosed Scope
    b = 2
    print(b)
    print(c)
    def inner():
        #Local Scope
        c = 3
        print(b)
        print(c)
    inner() #doesn't matter that we've called inner()

outer()
```

```python
#Built-in Scope (def)
def outer():
    #Enclosed Scope
    b = 2
    print(b)
    #print(c)
    def inner():
        #Local Scope
        c = 3
        print(b)
        print(c)
    inner()
outer()
```

As you can see, you can access b from the local and enclosed scope but you can not access c from the enclosed scope as it is a part of the local scope.
